```python
from lifelines import CoxPHFitter
import pandas as pd
from survival_utils import normalize_datasets
event = 'PFI'
cancer_type = ['BRCA']
feats_list = pd.read_csv('noncorrolated_feature_list.csv', header=None)[0].to_list()
feats = feats_list
df = pd.read_csv('/mnt/gpfs01/lsf-workspace/u2070124/Data/Data/pancancer/tcga_features_clinical_merged.csv')
df = df.loc[df['type'].isin(cancer_type)]
df = df[[event, f'{event}.time']+feats]
df = df.dropna()
df, _ = normalize_datasets(df, df, df.columns[2:], norm_type='meanstd')
cph = CoxPHFitter()
cph.fit(df, f'{event}.time', event, robust=True)
cph.print_summary(decimals=3, columns=['exp(coef)', 'exp(coef) lower 95%', 'exp(coef) upper 95%', 'p'])
```
