```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


def calculate_and_plot_correlation_normalized(df, col1_name, col2_name):
  """Calculates and plots the correlation between two columns of a pandas DataFrame after normalizing the values.

  Args:
      df: The pandas DataFrame.
      col1_name: The name of the first column.
      col2_name: The name of the second column.
  """

  # Normalize the data
  scaler = StandardScaler()
  normalized_data = scaler.fit_transform(df[[col1_name, col2_name]])

  # Create a new DataFrame with the normalized data
  normalized_df = pd.DataFrame(normalized_data, columns=[col1_name, col2_name])

  # Calculate the correlation coefficient
  correlation = normalized_df[col1_name].corr(normalized_df[col2_name])

  # Create the plot
  sns.scatterplot(x=col1_name, y=col2_name, data=normalized_df)
  # plt.title(f'between {col1_name} and {col2_name} : {correlation:.2f}')
  plt.title(f'{correlation:.2f}')
  plt.show()

df = pd.read_csv("/mnt/gpfs01/lsf-workspace/u2070124/Data/Data/pancancer/tcga_features_final.csv")
# df = df[df["type"]=="BRCA"]
calculate_and_plot_correlation_normalized(df, "mit_nodeDegrees_mean", "mit_assortCoeff")
```
