```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Function to create a random graph for a city
def create_city_graph(num_nodes, edge_prob=0.1):
    # G = nx.erdos_renyi_graph(n=num_nodes, p=edge_prob, seed=42)
    G = nx.random_geometric_graph(num_nodes, edge_prob, dim=2)
    return G

# Function to visualize the graph with nodes colored by normalized harmonic centrality
def visualize_graph(graph, node_colors, title):
    plt.figure(figsize=(8, 8))
    pos = nx.spring_layout(graph, seed=42)  # Layout for consistent node positioning
    nodes = nx.draw_networkx_nodes(graph, pos, node_color=node_colors, cmap=plt.cm.viridis, 
                                   node_size=300, alpha=0.8, vmin=0.25, vmax=0.5)
    edges = nx.draw_networkx_edges(graph, pos, alpha=0.5)
    nx.draw_networkx_labels(graph, pos, font_size=10, font_color="white")
    
    plt.colorbar(nodes)
    plt.title(title)
    plt.show()

# Create graphs for City A (100 nodes) and City B (50 nodes)
N_a = 100
N_b = 50
city_a = create_city_graph(N_a, edge_prob=0.2)
city_b = create_city_graph(N_b, edge_prob=0.2)

# Calculate harmonic centrality
harmonic_centrality_a = nx.harmonic_centrality(city_a)
harmonic_centrality_b = nx.harmonic_centrality(city_b)

# Normalize by maximum value
max_hc_a = max(harmonic_centrality_a.values())
max_hc_b = max(harmonic_centrality_b.values())

normalized_hc_a = {node: hc / (N_a-1) for node, hc in harmonic_centrality_a.items()}
normalized_hc_b = {node: hc / (N_b-1) for node, hc in harmonic_centrality_b.items()}

# Visualize graphs with nodes colored by normalized harmonic centrality
# Convert the normalized centrality values to a list of colors for visualization
node_colors_a = list(normalized_hc_a.values())
node_colors_b = list(normalized_hc_b.values())

visualize_graph(city_a, node_colors_a, "City A: Normalized Harmonic Centrality")
visualize_graph(city_b, node_colors_b, "City B: Normalized Harmonic Centrality")
```
